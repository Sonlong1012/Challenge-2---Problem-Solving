```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

# Initialize the inventory
inventory = [
    Product("Laptop", 999.99, 5),
    Product("Smartphone", 499.99, 10),
    Product("Tablet", 299.99, 9),
    Product("Smartwatch", 199.99, 3),
]

# Task 1: Calculate total inventory value
def calculate_total_value(products):
    total_value = sum(product.price * product.quantity for product in products)
    return total_value

# Task 2: Find the most expensive product
def find_most_expensive(products):
    most_expensive = max(products, key=lambda p: p.price)
    return most_expensive.name

# Task 3: Check if "Headphones" is in stock
def is_in_stock(products, product_name):
    return any(product.name == product_name for product in products)

# Task 4: Sort products in descending order by price and quantity
def sort_products(products):
    sorted_by_price = sorted(products, key=lambda p: p.price, reverse=True)
    sorted_by_quantity = sorted(products, key=lambda p: p.quantity, reverse=True)
    return sorted_by_price, sorted_by_quantity

# Output results
total_value = calculate_total_value(inventory)
most_expensive_product = find_most_expensive(inventory)
headphones_in_stock = is_in_stock(inventory, "Headphones")
sorted_by_price, sorted_by_quantity = sort_products(inventory)

# Display outputs
print(f"Total Inventory Value: {total_value:.2f}")
print(f"Most Expensive Product: {most_expensive_product}")
print(f"Are 'Headphones' in stock? {'Yes' if headphones_in_stock else 'No'}")
print("\nProducts sorted by price (descending):")
for product in sorted_by_price:
    print(f"{product.name}: ${product.price:.2f}, Quantity: {product.quantity}")

print("\nProducts sorted by quantity (descending):")
for product in sorted_by_quantity:
    print(f"{product.name}: ${product.price:.2f}, Quantity: {product.quantity}")
```
